## Continuous Numerical Variables:

### Positive Linear Relationship 

#### showing the correlation between **price and engine size**

```python
sns.regplot(x='engine-size', y='price', data=df) 
plt.ylim(0,) # set the y axis lower limit 
plt.show()
```

```python
df[["engine-size", "price"]].corr()
```

#### showing the correlaiton between **price and highway-mpg**

```python
df[["price","highway-mpg"]].corr()
```

```python
sns.regplot(x="highway-mpg", y="price", data=df)
plt.ylim(0,)
plt.show()
```

### Weak Linear Relationship

#### showing the correlaiton between **price and peak-rpm**

```python
sns.regplot(x="peak-rpm", y="price", data = df) 
plt.ylim(-10000,)
plt.show()
```

## Categorical Variables

These are variables that describe a 'characteristic' of a data unit, and are selected from a small group of categories. The categorical variables can have the type **"object" or "int64"**. A good way to visualize categorical variables is by using **boxplots.**

### Drawing plots by establishing correlaiton

```python
sns.boxplot(x="body-style", y="price", data=df)
plt.ylim(0,)
plt.show()
```

```python
sns.boxplot(x="engine-location", y="price", data=df)
plt.show()
```

```python
sns.boxplot(x="drive-wheels", y="price", data=df)
plt.ylim(0,)
plt.show() 
```

## Descriptive Statistical Analysis

### Value Counts

```python
df['body-style'].value_counts().to_frame()
```

```python
df['drive-wheels'].value_counts()
```

```python
df['drive-wheels'].value_counts().to_frame()
```

Let's repeat the above steps but save the results to the dataframe "drive_wheels_counts" and rename the column 'drive-wheels' to 'value_counts'.

```python
drive_wheels_count = df['drive-wheels'].value_counts().to_frame()
drive_wheels_count.reset_index(inplace=True)
drive_wheels_count = drive_wheels_count.rename(columns={"wheels":"counts"})
drive_wheels_count
```

```python
drive_wheels_count.index.name = 'drive-wheels'
drive_wheels_count
```

```python
engine_location_counts = df['engine-location'].value_counts().to_frame() 
engine_location_counts.rename(columns={"location":"counts"}, inplace=True)
engine_location_counts.index.name="eng_location"
engine_location_counts.head(10)
```
