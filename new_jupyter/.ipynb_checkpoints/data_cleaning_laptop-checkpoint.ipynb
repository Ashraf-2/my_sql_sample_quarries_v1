```python
df[['Screen_Size_cm']] = np.round(df[['Screen_Size_cm']],2)
df.head()
```

### Evaluate Missing data 

```python
missing_data = df.isnull() 
# print(missing_data.head())
```

```python
missing_data
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

### Replace the missing data of the most frequent value 

Hint: use mean value

```python
df
```

```python
screen_mean = df['Screen_Size_cm'].mean()
screen_mean
```

```python
df['Screen_Size_cm'].replace(np.nan,screen_mean, inplace=True)
```

```python
df
```

##### Replace the "weight_kg"'s missing values by the most frequent value of the column

```python
missing_weight = df['Weight_kg'].mean()
missing_weight
```

```python
common_weight = df['Weight_kg'].value_counts().idxmax()
common_weight
```

```python
df['Weight_kg'].replace(np.nan,common_weight, inplace=True)
```
