## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
x = df[['Experience Years']]
x
```

```python
y = df[['Salary']]
y
```

```python
print(x.mean())
```

```python
print(y.mean())
```

```python
plt.scatter(x.mean(), y.mean(), color="red") 
plt.scatter(x,y)
plt.xlabel("Experiencce (yrs)")
plt.ylabel("Salary")
plt.show()
```

##### here comes to the actual part of Linear Regression 

#### Train the model 
fit the model 

```python
reg.fit(x,y)
```

```python
m = reg.coef_
c = reg.intercept_
```

```python
print(m,c)
```

```python
y = m * x + c
y
```

```python
reg.predict(x) # your exact model it is! 
```

### let's create a new column of predicted salary using your model 

```python
df['Predicted_salary'] = reg.predict(x)
df
```

```python
# now draw a plot using your predicted salary for the experience
#plt.scatter(x,y)
plt.plot(x, df['Predicted_salary'])
plt.scatter(x.mean(), y.mean(), color="red")
```

## Count the Cost and Lost 

Recidual = cost and Lost

```python
df['lost'] = df['Salary'] - df['Predicted_salary']
df.head()
```

#### MSE and MAE 

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse = mean_squared_error(df['Salary'], df['Predicted_salary'])
mse
```
