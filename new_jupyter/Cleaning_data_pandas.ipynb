### Separating Columns 

```python
import pandas as pd

# Data for the DataFrame
data = {
    "name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "id": [101, 102, 103, 104, 105],
    "age": [25, 30, 22, 35, 28],
    "color": ["Red", "Blue", "Green", "Yellow", "Black"],
    "phone_number": ["123-456-7890", "234-567-8901", "345-678-9012", "456-789-0123", "567-890-1234"],
    "address": [
        "12345 Elm St, Springfield, IL",
        "67890 Oak Ave, Lincoln, NE",
        "11223 Pine Rd, Austin, TX",
        "44556 Maple Blvd, Denver, CO",
        "77889 Cedar Ln, Salem, OR"
    ]
}

# Create the DataFrame
street_df = pd.DataFrame(data)
street_df
```

```python
splitted_address =  street_df['address'].str.split(',',n=2,expand=True)
splitted_address
```

```python
street_df[["street_address","district","state"]] = street_df['address'].str.split(',',n=2, expand=True)
```
