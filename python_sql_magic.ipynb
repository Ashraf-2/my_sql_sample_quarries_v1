## Using Python Variables in your SQL Statements

You can use python variables in your SQL statements by adding a ":" prefix to your python variable names.
For example, if I have a python variable country with a value of "Canada", I can use this variable in a SQL query to find all the rows of students from Canada.

```python
country = "canada" 
%sql select * from std_test_score;
```

```python
%sql select distinct(country) from std_test_score order by country;
```

## You can use any python variable in your sql quarry. 
To use the variable you need to declare the variable like this 
c1 = "Hello";
and in SQL command, you have to put ':' this semicolon infront of the variable.

```python
c1 = "Russia"
%sql select * from std_test_score  where country = :c1;
```
